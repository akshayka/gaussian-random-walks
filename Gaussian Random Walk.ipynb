```julia
using Distributions
```

```julia
using PyPlot
```

```julia
N = 1000
TRIALS = 10000
MU = 0
SIGMA = 1
```

```julia
data = []
for i = 1:TRIALS
    mu = MU
    for j = 1:N
        mu = rand(Normal(mu, SIGMA))
    end
    push!(data, mu)
end

figure()
title("Shifting Normals Marginal Distribution")
plt[:hist](data, 100)
```

```julia
data = []
for i = 1:TRIALS
    sample = sum(rand(Normal(MU, SIGMA), N))
    push!(data, sample)
end

figure()
title("Sum of Normal Distributions")
plt[:hist](data, 100)
```

```julia
data = []
for i = 1:TRIALS
    sample = rand(Normal(MU, sqrt((SIGMA^2) * N)))
    push!(data, sample)
end

figure()
title("Sum of Normals")
plt[:hist](data, 100)
```
